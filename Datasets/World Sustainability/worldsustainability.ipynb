```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('WorldSustainabilityDataset.csv')
df.head()
```

```python
rename_columns=df.columns.str.split('(').str[0]
rename_columns=rename_columns.str.split('-').str[0]
```

```python
#Replace the columns with the new names
df.columns=rename_columns
df.head()
```

```python
#Missing values
missing_values=df.isnull().sum()
missing_values
```

```python
#Get the columns with dtype object
df.select_dtypes(include=['object']).columns

for i in df.select_dtypes(include=['object']).columns:
    df[i]=df[i].fillna(df[i].mode()[0])


```

```python
df.isnull().sum()

#Get the columns with dtype object
df.select_dtypes(include=['float']).columns

for i in df.select_dtypes(include=['float']).columns:
    df[i]=df[i].fillna(df[i].mean())


df.isnull().sum()
```

```python
df.to_csv('WorldSustainabilityDataset_cleaned.csv',index=False)
```
