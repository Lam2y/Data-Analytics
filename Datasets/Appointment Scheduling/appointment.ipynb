```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Appointment Scheduling Cleaned Data.csv')
```

```python
df.head()
```

```python
waiting_time = []
for i in df['Waiting Time'].str.split(','):
    waiting_time.append(str(i[0]).strip('days'))
    
    
for i in range(len(df['Waiting Time'])):
    df['Waiting Time'][i] = waiting_time[i]

print(df)

```

```python
df.rename(columns={"Waiting Time": "Waiting Time (days)"})
```
