```python
df_agg = df.groupby(['Year','Country'])[['GDP (in Trillions USD)']].sum()
df_agg

df_unstack = df_agg.unstack(level='Country').fillna('Unknown')
year = df_unstack.index
GDP = [df_unstack['GDP (in Trillions USD)',num] for num in df_unstack['GDP (in Trillions USD)'].columns]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.stackplot(year, GDP, labels=df_agg.unstack(level='Country'))
plt.show()
```

```python
df_agg = df.groupby('Year')[['GDP (in Trillions USD)']].mean()
plt.figure(figsize=(10,8))
plt.errorbar(df_agg.index, df_agg['GDP (in Trillions USD)'], yerr=0.05*df_agg['GDP (in Trillions USD)'])
plt.show()
```
